Bazen çok sayıda tekrar eden işlem yaparız. **"Sürekli tekrarla"** komutumuzla aynı anlama gelen işlemlere döngüler deriz. Bizim döngülerimiz bazen 0'dan 100'e kadar saymayı da kolaylıkla yapmamızı sağlar.

İki tane döngü şeklimiz var: **For Döngüsü** ve **While Döngüsü**


```python
#Bir For döngüsü oluşturalım.
for i in range(1,11):
  print(i)
#For döngüsünün yazımı aslında çok basit:
#for ..... in ..... :
#for, in ve iki nokta üst üste her zaman sabittir.
#Yukarıdaki i bir değişkendir ve for döngüsünde saymak için yer alır.
# in yazısının sağındaki yazılansa i'nin içinde tek tek yazdığı büyük veridir.
#Bu büyük beri bazen bir liste bazen bir değişken olablir.
#Yukarıdaki örnekte range dediğimiz şey bir aralktır.
#Koddaki "range" kodu 1 ila 10 arasındaki sayıları saymayı sağlıyor.
#Her zaman aralıkta yazılanınbir eksiğini ekrana yazar.
```

#While Döngüsü
While döngüsü, for'dan farklı olarak bir koşulun oluşması durumunda çalışır.
Bu koşul sayıların dışında mantıksal koşulları da çalıştırır. En temelinde if ve for'un karışımı gibidir.<br>
Örneğin:

1. Eğer a'nın değeri 40'tan küçük olduğu sürece çalış,
2. Eğer parola Doğru olduğu müddetçe/sürece çalış,
3. Eğer kullanıcı_adı = "Kratos The Teacher" olduğu sürece çalış.

While döngüsü, her zaman bir koşula ihtiyaç duyar!

```python
bilgi = (input("Kullanıcı adınızı giriniz: "))
while(bilgi != "Zeynep") or (bilgi != "zeynep"):
  print("Kullanıcı adı yanlış!")
  bilgi = (input("Kullanıcı adınızı giriniz: "))
print("Tebrikler, doğru kullanıcı adı girdiniz!")
```

#NOT
Asla sonsuz döngü kullanmayın! Yoksa program asla döngüden çıkamaz!

```python
toplam = 0
while(True):
  print("Çıkmak için 0 yazınız.")
  bilgi = int(input("Bir sayı giriniz: "))
  toplam += bilgi #toplam = toplam + bilgi ile aynıdır.
  if(bilgi == 0):
    break #Döngüyü kırmak için kullanılır.
print("Girdiğiniz sayıların toplamı: ",toplam)
```

#NOT
While döngüsünü kırmak için **"break"** yani kırmak anlamına gelen komutu kullanırız. Eğer bir döngü veya if koşulunu pas geçip içi boş kalsın istersek **"pass"** komutunu kullanırız.

```python
sayim = int(input("Lütfen sayı giriniz: "))
if sayim % 2 == 0:
  pass
if sayim % 4 == 0:
  print("Sayı 4'e bölünür.")
```

```python
#Sayı tahmin oyunu
import random
rastgele_sayi = 0
yas = int(input("Lütfen yaşınızı giriniz: "))
if(yas >= 1 and yas >= 10):
  print("Rastgele sayı aralığınız 1-20 arasındadır.")
  rastgele_sayi = random.randint(1,20)
if(yas >= 11 and yas >= 30):
  print("Rastgele sayı aralığınız 1-50 arasındadır.")
  rastgele_sayi = random.randint(1,50)
if(yas >= 31 and yas >= 51):
  print("Rastgele sayı aralığınız 1-100 arasındadır.")
  rastgele_sayi = random.randint(1,100)

#Ana oyun  döngüsü
tahmin_hakki = 5
tahmin = int(input("Lütfen tahmininizi giriniz: "))
while(tahmin != rastgele_sayi):
  tahmin_hakki -= 1
  print("Kalan tahmin hakkınız:",tahmin_hakki)
  tahmin = int(input("Lütfen tahmininizi giriniz: "))
  if(tahmin_hakki == 0):
    print("Bilemediniz! Bilgisayarın tuttuğu sayı:",rastgele_sayi)
    break
  if(tahmin == rastgele_sayi):
    print("Tebrikler, doğru tahmin!")
    break
```

#For Döngüsü ile Yıldız Çizme

```python
for k in range(1,10,1):
  for i in range(k):
    print("*", end=" ")
  print()

for k in range(10,-10,-1):
  for i in range(k):
    print("*", end=" ")
  print()
```
