##Fonksiyona Parametre Yazma
Fonksiyonda çalışacak değişkenlere parametreler denir. Bu parametreler fonksiyonun içinde çalışacak değerlerdir. Ana programda yazılımcı zorunlu olarak bunu doldurmalıdır. İster doğrudan bir değer yazar isterse elle bilgi girer.

````python
#Parametre parantez içine yazılır.
def kullanici_paneli(kullanici_adi, parola):
  if kullanici_adi == "Gary" and parola == "12A34B":
    print("Başarılı kullanıcı girişi!")
  else:
    print("Başarısız Giriş!")
````

````python

#Değeri doğrudan yazarsam çalışır.
kullanici_paneli("Gary", "12A34B")
#Input yazarak çalışır.
kullanici_ad = input("Lütfen kullanıcı adınızı yazınız:")
sifre = input("Lütfen şifrenizi giriniz:")
kullanici_paneli(kullanici_ad,sifre)
````

#Değer Döndüren ve Değer Döndürmeyen Fonksiyonlar

Bizim oluşturduğumuz daha önceki fonksiyonlar **değer döndürmeyen** fonksiyonlardı. Yani işlemi yaptırır ve sadece bir kod satırında çağırarak
çalışırlardı. Ancak **değer döndüren** fonksiyonlar ise belirli bir tür değişkenin içerisine veri aktarılarak çalışan fonksiyonlardır.

##Peki, Değer Döndüren Fonksiyon ne demek oluyor?
Değer döndürmeyen fonksiyon bir değişkenin içerisine değer atılarak yazılır.Bu değişken fonksiyonda belirlenen aynı türden değişken olmalıdır. Yani bir fonksiyon sonucunda değişkene sadece bir değer atılır. Bu değer string ise string türünde, integer ise integer türünde bir değişkenin içine yazılır.

````python
#Değer Döndürmeyen fonksiyon
def toplam  (sayi1,sayi2):
  return sayi1 + sayi2 #Return komutunu kullanıyorum.

#ANA PROGRAM
a = int(input("Birinci sayıyı giriniz:"))
b = int(input("İkinci sayıyı giriniz:"))
sonuc = toplam(a,b) #Bir değişken içerisine yazıyorum.
print(sonuc)
````

````python
# 4 işlemi değer döndüren fonksiyon ile yapın her bir işlem ayrı bir fonksiyon
def islem(sayi1, sayi2):
  return sayi1 + sayi2
def islem2(sayi1, sayi2):
  return sayi1 - sayi2
def islem3(sayi1, sayi2):
  return sayi1 * sayi2
def islem4(sayi1, sayi2):
  return sayi1 / sayi2
a = int(input("Lütfen birinci sayıyı giriniz:"))
b = int(input("Lütfen ikinci sayıyı giriniz:"))
print("Toplama:", islem(a,b))
print("Çıkarma:", islem2(a,b))
print("Çarpma:", islem3(a,b))
print("Bölme:", islem4(a,b))
````

````python
import random

#1. Kendi kağıdımı oluşturuyorum -> bir liste
kullanici_listesi = []
#2. Makinenin kağıdını oluşturuyorum -> bir liste
rastgele_listesi = []
#3. Kağıdımı dolduruyorum 6 rakamla
for liste_olusturma in range(0,6):
    sayim = int(input("1-50 arasında sayı giriniz: "))
    kullanici_listesi.append(sayim)
#4. Makinenin listesini dolduruyorum.
for liste_doldurma in range(0,6):
    sayim = random.randint(1,50)
    rastgele_listesi.append(sayim)
#5. Kendi listem ve rastgele liste arasında elemanları sırayla karşılaştırma
for c in range(0,6):
  if kullanici_listesi[c] == rastgele_listesi[c]:
   print("Bir eşleşme gerçekleşti")
  print(kullanici_listesi[c] ,"=", rastgele_listesi[c])
print(kullanici_listesi)
print(rastgele_listesi)
````

#OS KÜTÜPHANESİ İLE DOSYA İŞLEMİ YAPMA

OS kütüphanesi "Operating System" yani "işletim Sistemi" anlamına gelmektedir. Dosya açma, oluşturma, klasör yollarını öğrenme, klasör oluşturma gibi işletim sisteminizde gezinmenizi sağlayan bir kütüphanedir.
os kütüphanesinde yaygın kullanılan open fonksiyonu vardır.

NOT: Oluşturacağınız dosyanın uzantısını bilmeniz gerekir.
Word dosyasının uzantısı .docx
Not defterinin uzantısı .txt
Örnek bir resim dosyası .png olabilir.

os.open -> Bir dosya aç demek. Bu dosya klasör veya herhangi bir dosya olabilir. <br>
w -> write yani dosyaya yaz demek. <br>
r -> read sadece oku demek. <br>
a -> all yani hem yaz hem oku.
