#Listeler
Değişkenler bir tane veri saklayan ve tek amaçla oluşturulan yapılardı.
Listeler ise birden fazla veriyi içinde saklayan ve farklı türden verileri kendi içinde tutabilen yapılardır.<br>

Liste oluşturmak oldukça basittir. Öncelikle listenin adı verilir ve eşittir işareti konulup köşeli parantezler arasına değerli [ ] yazılır.

```python
sehirler = ["Bursa","Ankara","İstanbul","İzmir"]
haftanin_gunleri = ["Pazartesi","Salı","Çarşamba","Perşembe","Cuma","Cumartesi","Pazar"]
rastgele_liste = ["Elma",12,"Bilinmez","Çorba"]
print(sehirler)
print(haftanin_gunleri)
print(rastgele_liste)
```

```python
maker = ["Müfit Führer","Zeynep kpop", "Ataberk Fener","Yigit suyun ph degeri","Teacher the kratos"]
print(maker[1])
```

**NOT:** Listelerde bir elemanı içinden çağırmak istiyorsak listenin adını yazıp yanına köşeli parantez [ ] açıp içerisine rakam yazıyoruz. Bu rakam 0'dan başlar ve listenin eleman sayısının 1 eksiğine kadar gider. Mesela haftanın günleri 7 ise listedeki elemanları 0 ve 6 arasında yazdırabilirim. <br>

Böylece ile eleman 0'dır. Bu köşeli parantez içindeki değerlere **listenin indisi** denir.

```python
import random as rd
listem = ["elma","armut","ayva",124,True]
rastgele_indis = rd.randint(0,len(listem)-1)
print(listem[rastgele_indis])
```

#Listelere İnput ile Eleman Ekleme ve Çıkarma
Input kullanarak listelere eleman ekleyip çıkarabiliriz. Bazı programlama dillerinde add (ekleme) veya append (listeye içine ekleme) anlamına gelen bir fonksiyon ile ekleme yapılabilir.<br>

Çıkarma işlemlerinnde delete (silme) veya remove (listeden çıkarma) anlamında bu fonksiyonlar kullanılabilir. İkisinden örnek vermek gerekirse liste.append(sayi) veya liste.remove(3) gibi farklı şekillerde kullanılabilir.

```python
tatlılar = ["kemalpaşa tatlısı","süt helvası","süt kadayıf","irmik helvası","tiramisu","cheesecake","muffin"]
tatli = input("Lütfen tatlı giriniz.")
tatlilar.append(tatli)
print(tatlilar)
#Listeden ekleneni silebilirim.
tatlilar.remove(tatli)
print("Silme işlemi")
print(tatlilar)
silinecek = input("Silinecek başka bir tatlı seçin.")
tatlilar.remove(silinecek)
print("Ekstra silinmeden sonra:")
print(tatlilar)
```

#Listelerde Farklı İşlemler
Listeleri temizleme, eleman seçip aynı anda çıkarma, listeyi tersten yazdırma,liste eleman sayısı veya indis bilgisi alma gbi işlemler yapabiliriz.

```python
#Clear komutu -> Her şeyi sildi.
haftanin_gunleri = ["Pazartesi","Salı","Çarşamba","Perşembe","Cuma","Cumartesi","Pazar"]
haftanin_gunleri.clear()
print(haftanin_gunleri)
```

```python
#Count komutu -> Yazılan Elemanı Sayar.
haftanin_gunleri = ["Pazartesi","Salı","Çarşamba","Perşembe","Cuma","Cumartesi","Pazar"]
gun = haftanin_gunleri.count("Pazartesi")
print(gun)
```

```python
#Index komutu -> Yazılan elemanın kaçıncı index olduğunu sayar.
haftanin_gunleri = ["Pazartesi","Salı","Çarşamba","Perşembe","Cuma","Cumartesi","Pazar"]
print(haftanin_gunleri.index("Çarşamba"))
```

```python
#Pop komutu -> listenin içinden bahsedilen index'li elemanları alır yazar ve listeden çıkarır.
haftanin_gunleri = ["Pazartesi","Salı","Çarşamba","Perşembe","Cuma","Cumartesi","Pazar"]
gun = haftanin_gunleri.pop(3)
print(gun)
print(haftanin_gunleri)
```

```python
#Reverse komutu -> Listeyi Ters Çevirir
haftanin_gunleri = ["Pazartesi","Salı","Çarşamba","Perşembe","Cuma","Cumartesi","Pazar"]
haftanin_gunleri.reverse()
print(haftanin_gunleri)
```

#LİSTELERLE ARAMA YAPMA ve I
Listeler içerisinde belirli koşullar sağlandığında yazdırılacak veya işe yarayacak bilgileri alabilmek için listelerde arama yapılmalıdır.
Listelerde arama yapmak için döngüler kullanılabilir.

```python
ucerli_liste = []
for a in range(0,250,3):
  ucerli_liste.append(a)
print(ucerli_liste)

# Bu listenin içinden 5'e bölünen sayıları ekrana yazdırma

for eleman in range (0,len(ucerli_liste)-1):
  if ucerli_liste[eleman] % 5 == 0:
    print(ucerli_liste[eleman])


```

```python
# 0 - 250 arasına 3'er 3'er sayıları yazdırma
ucerli_liste = []
onbesin_kati = []
for a in range(0,250,3):
  ucerli_liste.append(a)
print(ucerli_liste)
print("Toplam eleman sayısı",len(ucerli_liste))
# Bu listenin içinden 5'e bölünen sayıları ekrana yazdırma
for eleman in range (0,len(ucerli_liste)-1):
  if ucerli_liste[eleman] % 5 == 0:
    onbesin_kati.append(ucerli_liste[eleman])
print(onbesin_kati)
print("Toplam eleman sayısı",len(onbesin_kati))
```
